```python
import pandas as pd

df2 = pd.read_csv("/content/sample_data/mnist_test.csv")

df2["0.1"] = df2["0.1"].astype(float)

for i in df2.index:
  if df2.loc[i, "0.1"] == 0:
    df2.loc[i, "0.1"] = i * 1.2
  elif df2.loc[i, "7"] == 1:
    df2.drop(i, inplace=True)

df2.head()
```
